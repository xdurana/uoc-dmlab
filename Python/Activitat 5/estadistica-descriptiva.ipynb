## Anàlisi univariant

Dins de l'anàlisi univariant, separarem variables numèriques i categòriques. Per les variables numèriques calcularem el número de valors disponibles, la mitjana, la desviació estàndard, el mínim, el màxim i els quartils, i n'analitzarem la seva distribució. Per les variables categòriques veurem la distribució dels seus valors en un histograma.

#### Test de normalitat

Si observem els valors mínim, mitjà i màxim veiem clarament que no és una variable amb distribució normal. En la majoria de models numèrics d'aprenentatge automàtic, necessitarem que les variables segueixin la distribució normal.

Amb la funció `distplot` de la llibreria de visualització `Seaborn` podem comparar la distribució de la nostra variable amb la normal.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

sns.histplot(df_train['SalePrice'], kde=False, stat='density')

mu, std = norm.fit(df_train['SalePrice'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plot the fitted normal distribution
plt.plot(x, p, 'k', linewidth=2)
plt.show()
```

D'una manera numèrica i menys visual podem calcular dos indicadors de normalitat. L'assimetria de la funció o _skewness_ i el _kurtosis_. 

La funció `skew` calcula el coeficient d'assimetria. Si és positiu i més gran que u, com en el nostre cas, els valors de la distribució són més densos cap a l'esquerra de la gràfica. En una distribució normal, el coeficient d'assimetria seria proper a zero.

```python
df_train['SalePrice'].skew()
```

La funció `kurtosis` ens dóna una idea de la relació del pic central amb els extrems de la campana de la distribució. Si és proper a 1, el valor serà coherent amb la normalitat de la variable. En el nostre cas, és ben lluny.

```python
df_train['SalePrice'].kurt()
```

Quan tenim una variable que no és normal, una de les transformacions clàssiques que pot funcionar és aplicar el logaritme a la variable.

```python
df_train['SalePrice'] = np.log1p(df_train['SalePrice'])

df_train['SalePrice'] = np.log1p(df_train['SalePrice'])

sns.histplot(df_train['SalePrice'], kde=False, stat='density')

mu, std = norm.fit(df_train['SalePrice'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plot the fitted normal distribution
plt.plot(x, p, 'k', linewidth=2)
plt.show()
```

Veiem com ara sí que té forma normal. Ho comprovem amb els mètodes estadístics per confirmar-ho.

```python
df_train['SalePrice'].skew()
```

```python
df_train['SalePrice'].kurt()
```

## Anàlisi bivariant

Per analitzar un conjunt de dades necessitem tenir en compte més d'una variable alhora. L'anàlisi bivariant permet identificar les relacions entre dues variables, i fins i tot veure de quina manera una pot predir l'altra. 

En aquest cas podem veure quina és la relació entre el preu de venda i la superfície construida amb un _scatter plot_ o diagrama de punts.

```python
var = 'GrLivArea'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice')
```

En el cas d'una variable categòrica com la qualificació, podem visualitzar la relació amb un _boxplot_ o diagrama de caixes.

```python
var = 'OverallQual'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
plt.xticks(rotation=90);
```

Si volem visualitzar alhora les relacions creuades entre vàries variables, podem fer un _pairplot_.

```python
sns.set()
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
sns.pairplot(df_train[cols], height = 2.5)
plt.show();
```

## Anàlisi multivariant

Finalment, ens interessarà veure quins conjunts de variables estan relacionats entre si. Per això, farem servir tècniques estadístiques d'anàlisi multivariant.

Una de les eines més útils és calcular la matriu de correlació entre les variables. Amb la funció `heatmap` podem visualitzar de manera fàcil aquelles variables més correlacionades, que corresponen a una intensitat major de color.

```python
corrmat = df_train.corr(method='spearman', numeric_only=True)
f, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corrmat, ax=ax, cmap="YlGnBu", linewidths=0.1)
```

El `clustermap` ens permet anar un pas més enllà i agrupar les variables que tenen més relació entre elles, a partir d'un algorisme de _clustering_ a partir de la informació de la correlació. Això ens servirà eventualment, per evitar escollir dades molt correlacionades que ens puguin perjudicar a l'hora de fer servir un model.

Veiem, per exemple, que les variables `YearBuilt`, `GarageYrBlt` i `YearRemodAdd` són molt correlacionades i ens dónen gairebé la mateixa informacio. En la fase de modelat, n'escolliriem només una de les tres.

```python
corrmat = df_train.corr(method='spearman', numeric_only=True)
cg = sns.clustermap(corrmat, cmap="YlGnBu", linewidths=0.1);
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
cg
```

Finalment, escollirem les 15 variables més correlacionades amb el preu de venda de l'habitatge, i en mostrarem la correlació.

```python
k = 15 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
f, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(cm, ax=ax, cmap="YlGnBu", linewidths=0.1, yticklabels=cols.values, xticklabels=cols.values)
```
