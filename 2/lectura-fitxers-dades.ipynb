# Introducció

En aquesta activitat veurem com llegir diferents tipus de fitxers (csv, excel, txt, json, xml i zip) i carregar-los en un DataFrame.

## Llegir un fitxer CSV

El fitxer que llegirem correspon al dataset de la competició [Titanic: Machine Learning from Disaster](https://www.kaggle.com/c/titanic) de [Kaggle](https://www.kaggle.com/).

[Loading A CSV Into pandas](https://chrisalbon.com/python/data_wrangling/pandas_dataframe_importing_csv/)

```python
import pandas as pd

df = pd.read_csv("data/titanic-train.csv")
df.head()
```

```python
df.to_csv('data/titanic.csv')
```

##  Llegir un fitxer Excel

[TMDB 5000 Movie Dataset](https://www.kaggle.com/tmdb/tmdb-movie-metadata)

[Reading and writing Excel files in Python pandas](https://medium.com/@kasiarachuta/reading-and-writingexcel-files-in-python-pandas-8f0da449cc48)

[Excel and Pandas](https://www.dataquest.io/blog/excel-and-pandas/)

```python
xls_file = pd.read_excel('data/movies.xls')
xls_file
```

```python
xls_file = pd.read_excel('data/movies.xls', sheet_name='2010s')
xls_file
```

```python
xls_file.to_excel('data/movies_2010.xlsx', index=False)
```

## Llegir un fitxer JSON

[Dictionary of American Regional English (DAREDS)](https://www.kaggle.com/rtatman/dictionary-of-american-regional-english-dareds/data)

[Pandas JSON](https://hayd.github.io/2013/pandas-json)

```python
import pandas as pd

df = pd.read_json('data/pandas_issues.json')
df.head(10)
```

```python
df.to_json('output/pandas_issues.json')
```

## Llegir un fitxer XML

[Ten thousand books, one million ratings. Also books marked to read, and tags](https://www.kaggle.com/zygmunt/goodbooks-10k)

```python
import xml.etree.ElementTree as ET

tree = ET.parse('data/sample_book.xml')
root = tree.getroot()
root.tag
```
